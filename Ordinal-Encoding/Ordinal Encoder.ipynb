```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("Ordinal.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df["Education"].unique()
```

```python
enc = OrdinalEncoder()
```

```python
print(enc)
```

```python
enc.fit_transform(df[["Education"]])
```

```python
df[["Education"]] = enc.fit_transform(df[["Education"]])
```

```python
df.head()
```

```python
education = ['PhD', 'Masters','Bachelors']
```

```python
enc = OrdinalEncoder(categories=[education])
```

```python
enc.fit_transform(df[["Education"]])
```

```python
df[["Education"]] = enc.fit_transform(df[["Education"]])
```
